```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.xlabel("Company",fontsize=10)
plt.ylabel("Revenue",fontsize=10)
plt.title("Company Vs Revenue")
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.legend()
plt.show()
```

```python
len(company)
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.xlabel("Company",fontsize=10)
plt.ylabel("Revenue",fontsize=10)
plt.title("Company Vs Revenue")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
#plt.show()
plt.savefig("histogram.jpg")
```

```python
plt.barh(xpos,revenue)
plt.show()
```

```python
np.random.seed(10)
data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]
```

```python
data
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(data,vert=False,patch_artist=True)
plt.show()
```

```python
#PIE CHART
# Draw pie chart to track down home expenses
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.figure(figsize=(10,7))
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct="%0.2f%%",shadow=True,explode=[0.1,0,0.2,0.1,0],startangle=180)
plt.legend(loc="best")
#plt.show()
plt.savefig("piechart_new.jpg")
```
