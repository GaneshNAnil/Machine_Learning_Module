```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_boston
from sklearn.preprocessing import StandardScaler
```

```python
boston=load_boston()
boston.data
boston_df=pd.DataFrame(boston.data, columns   = boston.feature_names)
```

```python
boston_df
```

```python
boston_df["Price"]=boston.target
```

```python
X=boston_df.drop(["Price"],axis=1)
y=boston_df["Price"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
dt_classifier = DecisionTreeRegressor()
dt_classifier.fit(X_train,y_train)
```

```python
dt_regressor = DecisionTreeRegressor()
dt_regressor.fit(X_train,y_train)

score = dt_regressor.score(X_train,y_train)
print("R-squared Train  :", score)

score = dt_regressor.score(X_test, y_test)
print("R-squared Test  :", score)

y_pred = dt_regressor.predict(X_test)

mse = mean_squared_error(y_test,y_pred)
print("MSE :",mse)
print("RMSE :",mse**(1/2.0))
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt_regressor,filled=True)
plt.show()
```

```python
dt_regressor = DecisionTreeRegressor(max_depth=5)
dt_regressor.fit(X_train,y_train)

score = dt_regressor.score(X_train,y_train)
print("R-squared Train  :", score)

score = dt_regressor.score(X_test, y_test)
print("R-squared Test  :", score)

y_pred = dt_regressor.predict(X_test)

mse = mean_squared_error(y_test,y_pred)
print("MSE :",mse)
print("RMSE :",mse**(1/2.0))
```
